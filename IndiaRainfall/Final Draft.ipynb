# Training the Model

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
y=df["ANNUAL"] #the column that needs to be predicted
X=df[["Jan-Feb","Mar-May","Jun-Sep","Oct-Dec"]] #Input data to predict
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42) #splitting the data for model training
```

```python
model=LinearRegression() #creating an instance of the LinearRegression() model
```

```python
model.fit(X_train,y_train) #Training the data
```

```python
preds=model.predict(X_test) #predicting the values from the test data and storing in preds
```

```python
preds #this is how predictions look like
```

### Metrics and Testing the Model

```python
from sklearn import metrics #importing the metrics library
```

```python
print(metrics.mean_absolute_error(y_test,preds)) #calculating the mean absolute error
print(metrics.mean_squared_error(y_test,preds)) #calculating the mean squared error
print(np.sqrt(metrics.mean_squared_error(y_test,preds)))c#calculating the RMSE value
print(metrics.r2_score(y_test,preds)) 
```

```python
plt.scatter(preds,y_test) #seeing if all the predictions are correctly matched or not
plt.xlabel("PREDICTIONS")
plt.ylabel("Y_TEST")
plt.tight_layout()
```
